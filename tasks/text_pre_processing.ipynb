```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv('tripadvisor_hotel_reviews.csv') # You can access an entire column by using the column name. E.g data['Rating']
data.info()
```

```python
data.head()
```

```python
data["Review"][0]
```

```python
# Create an extra column for lowercase version of each review
data['Lowercase_Review'] = data['Review'].str.lower()
data.head()
```

```python
# Remove stopwords from reviews
en_stop_words = stopwords.words("english")
en_stop_words.remove("not")  
data["No_Stopwords_Review"] = data["Lowercase_Review"].apply(
    lambda x: " ".join([word for word in x.split() if word not in en_stop_words])
)

"""
A lambda is just a short one-line function without a name.

Example:
lambda x: x + 1

is like writing:
def add_one(x):
    return x + 1

2. apply()
This is a Pandas method used to apply a function to every value in a column.

Example:
data['Age'].apply(lambda x: x + 1)

where x is the current row in the 'Age' column.
"""
data.head()
```

```python
data["No_Stopwords_Review"][0]
```

```python
# Remove punctuation from reviews

# A slightly different approach where we are accessing entire data frame and accesing one row at a time(axis =1). We then specify the column we need
# It's great for sceanrios where you want to apply a function to multiple columns in a row

# Replace * with star
data["No_Punctuation_Review"] = data.apply(
    lambda x: re.sub(r"[*]", "star", x["No_Stopwords_Review"]), axis=1
)

# Remove punctuation
data["No_Punctuation_Review"] = data.apply(
    lambda x: re.sub(r"[^\w\s]", "", x["No_Punctuation_Review"]), axis=1
)

data.head()
```

```python
# Tokenize the reviews
data["Tokenized_Review"] = data["No_Punctuation_Review"].apply(
    lambda x: word_tokenize(x) if isinstance(x, str) else []
)
data.head()
```

```python
# Stemming the reviews
ps = PorterStemmer()
data["Stemmed_Review"] = data["Tokenized_Review"].apply(
    lambda x: [ps.stem(word) for word in x]
)
data.head()
```

```python
# Lemmatization of the reviews
lemmatizer = WordNetLemmatizer()
data["Lemmatized_Review"] = data["Tokenized_Review"].apply(
    lambda x: [lemmatizer.lemmatize(word) for word in x]
)
data.head()
```

```python
# simple exploration of the lemmatized tokens
tokens_clean = sum(data["Lemmatized_Review"], []) # Flatten the list of lists into a single list - one big list of tokens
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())

print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigrams)
```
